```python
f = open('CONSD.txt', 'r')
data = []
for line in f:
    data.append(line)
data
```

```python
norm_data = []
last_stroke = ''
for i in data:
    q = list(i)
    if q[0] == '>':
        norm_data.append(last_stroke)
        last_stroke = ''
        continue
    b = ''.join(q[:(len(q)-1):])
    last_stroke += b
norm_data.append(last_stroke)
norm_data = norm_data[1::]
norm_data
```

```python
lent = len(norm_data[0])
a = []
c = []
g = []
t = []
tr_stroke = ''
for i in range(0, lent):
    a.append(0)
    c.append(0)
    g.append(0)
    t.append(0)
    for j in norm_data:
        if (j[i] == 'A'):
            a[i] += 1
        if (j[i] == 'T'):
            t[i] += 1
        if (j[i] == 'C'):
            c[i] += 1
        if (j[i] == 'G'):
            g[i] += 1
    if ((a[i] >= c[i]) & (a[i] >= g[i]) & (a[i] >= t[i])):
        tr_stroke += 'A'
    elif ((c[i] >= a[i]) & (c[i] >= g[i]) & (c[i] >= t[i])):
        tr_stroke += 'C'
    elif ((g[i] >= a[i]) & (g[i] >= c[i]) & (g[i] >= t[i])):
        tr_stroke += 'G'
    elif ((t[i] >= a[i]) & (t[i] >= g[i]) & (t[i] >= c[i])):
        tr_stroke += 'T'
```

```python
print('A:', *a)
print('C:', *c)
print('G:', *g)
print('T:', *t)
print(tr_stroke)
```
