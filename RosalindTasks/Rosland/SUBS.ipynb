```python
f = open('SUBS.txt', 'r')
data = []
location = []
for line in f:
    data.append(line)
leng = len(data[1])
f.close()
for i in range(0, len(data[0])):
    if (data[0][i] == data[1][0]):
        flag = 0
        for j in range(1, leng - 1):
            if(data[0][i + j] != data[1][j]):
                flag = 1
                break
        if(flag == 0):
            location.append(i)
location = [(x + 1) for x in location]
print(*location)
```

```python
GCcont = 0
num_max = 0
count = 0
b_max = 0
for i in data[1::2]:
    Gc = 0
    b = len(i)
    for j in i:
        if((j == 'G') or (j == 'C')):
            Gc += 1
    if((Gc/b) > GCcont):
        GCcont = (Gc)
        num_max = count
        b_max = b
    count += 1
print(b_max)
print(GCcont)
print(data[2*num_max])
```
