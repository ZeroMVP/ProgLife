```python
f = open('rosalind_gc.txt', 'r')
raw_data = []
for i in f:
    raw_data.append(i)
```

```python
norm_data = []
for i in raw_data:
    q = list(i)
    if q[0] == '>':
        norm_data.append(i)
        continue
    b = ''.join(q)
    norm_data.append(b)
norm_data
```

```python
top = ''
topGC = 0
last = ''
last_len = 0
GC = 0
Y_num = 0
for i in norm_data[1::]:
    temp = list(i)
    if (temp[0] == '>'):
        if ((GC / last_len) > topGC):
            top = last
            topGC = GC / last_len
        GC = 0
        last_len = 0
        last = i
        continue
    else:
        for i in temp:
            if(i == 'G' or i == 'C'):
                GC += 1
        last_len += len(temp) - 1
print(top, topGC)
```
